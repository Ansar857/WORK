```python
margen_10=[]
margen_6=[]
no_margen=[]
```

```python
for i,b, p,a,pa in zip(id_1,buybox, price,asian,pack):
    p=float(p)
    pa=float(pa)
    pack=p*pa
    #print(b,p)
    #print(pack)"""
    try_new_product=driver.find_element_by_xpath('//*[@id="search-again"]/span/input').click()
    time.sleep(3)
    clear=driver.find_element_by_xpath('//*[@id="search-string"]').clear()
    time.sleep(3)
    ids=driver.find_element_by_xpath('//*[@id="search-string"]')
    ids.send_keys(a)
    #search
    driver.find_element_by_xpath('//*[@id="a-autoid-1"]/span/input').click()
    #sale rate
    time.sleep(3)
    ids=driver.find_element_by_xpath('//*[@id="afn-pricing"]')
    ids.send_keys(b)
    ids=driver.find_element_by_xpath('//*[@id="afn-fees-inbound-delivery"]')
    ids.send_keys('.50')
    ids=driver.find_element_by_xpath('//*[@id="afn-cost-of-goods"]')
    ids.send_keys(pack)
    time.sleep(3)
    driver.find_element_by_xpath('//*[@id="update-fees-link-announce"]').click()
    time.sleep(3)

    revinue=driver.find_element_by_xpath('//*[@id="bar2"]/span/span').text
    time.sleep(3)

    profit=driver.find_element_by_xpath('//*[@id="bar8"]/span/span').text
    time.sleep(3)

    r=float(revinue)
    p=float(profit)
    net_profit=(p/r)*100
    if net_profit > 10.00:
        li = [i,a,b,pack,net_profit]
        margen_10.append(li)
    else:
        li3=[i,a,b,pack,net_profit]
        no_margen.append(li3)
```
